# Scrape Jobs Data From WUZZUF Plateforme

### The Map is to get : - job title - company name - job type - experience level

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
```

## input url

```python
url = 'https://wuzzuf.net/search/jobs/?a=hpb&q=data%20analyst&start=1'
```

## Create client-based request to get the URL

```python
client = urlopen(url)
```

## Getting the html source of all the page

```python
html = client.read()
```

## Close the request

```python
client.close()
```

## Create html parser Using BS4

```python
soup = BeautifulSoup(html,'html.parser')
soup
```

## Getting the data needed

```python
jobs = soup.find_all("div", {'class':'css-1gatmva e1v1l3u10'})
df = []
for job in jobs:
    jobs_description = {
        'title':job.find('h2', class_='css-m604qf').text,
        'company' : job.find('a', class_='css-17s97q8').text,
        'job_type' : job.find('span',class_='css-1ve4b75 eoyjyou0').text,
        'published_date' : job.find('div',class_='css-do6t5g').text,
        'location' : job.find('span',class_='css-5wys0k').text,
        'job_url' : 'https://wuzzuf.net'+job.find('a', class_='css-o171kl')['href']
    }
    df.append(jobs_description)
print(df)
    
```

## Return the data into  DataFrame

```python
data = pd.DataFrame(df)
data
```

## Return the data into Excel file

```python
data.to_excel('WUZZUF_JOBS.xlsx')
```
